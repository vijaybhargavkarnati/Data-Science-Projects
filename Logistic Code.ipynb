```python
+## Data Wranding libs
import pandas as pd
import numpy as np
import math

## Modeling libs
from sklearn.linear_model import LinearRegression
from sklearn.metrics import confusion_matrix, accuracy_score
```

<h1 style='color: rgb(0, 56, 150)'> Dataset Creation </h1>

```python
np.random.seed(2)
x1 = np.random.rand(20)
x2 = np.random.rand(20)*100*np.random.rand(20)
y = np.random.randint(0, 2, 20)
```

```python
df = pd.DataFrame({'x1': x1, 'x2': x2, 'y': y})
```

```python
df.head()
```

<h1 style='color: red'> Building classifier from linear regression </h1>

```python
X = df[['x1', 'x2']]
y = df[['y']]
```

```python
linearRegCal = LinearRegression()
linearRegCalFit = linearRegCal.fit(X, y)
```

```python
dfAnalysis = df.copy()
dfAnalysis['linearPredRaw'] = linearRegCalFit.predict(X)
```

```python
dfAnalysis['linearPredRaw']
```

```python
dfAnalysis['linerPredFlag'] = [1 if i >= 0.6 else 0 for i in dfAnalysis['linearPredRaw']]
```

```python
dfAnalysis.head()
```

```python
print('*'*20)
print('Accuracy of simple linear regression on classification %s ' %(accuracy_score(y, dfAnalysis['linerPredFlag'])*100))
print('*'*20)
```
