We should take care of whethet our data is loaded properly or not. If not, we need perform data casting

```python
def ObjToFloat(colList , df):
        dfCopy = df.copy()
        
        for i in colList:
            dfCopy[i] =  pd.to_numeric(dfCopy[i])
        return  dfCopy
wrongcast = ['normalized-losses','bore','stroke','horsepower', 'peak-rpm','price']    
    
auto = ObjToFloat(wrongcast, auto)
```

Now the data is perfectly casted

<h5> Displaying the percentage of missing values for each feature in the data  </h5>

```python
auto.isnull().sum()
```

```python
def missing_Percentage(DataFrame):
    total = auto.isnull().sum().sort_values(ascending  = True)
    percent = round(auto.isnull().sum().sort_values(ascending  = True)/len(auto)*100,2)
    return pd.concat([total, percent], axis =1, keys=['Total','Percent'])

missing_percentage = missing_Percentage(auto)


#axis is used to display the percent values in a column adjacent to Total, if not given it will print horizontally
```

```python
missing_percentage
```
